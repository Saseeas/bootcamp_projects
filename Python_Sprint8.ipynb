```python
#filter with many condition   # and  &     , or | 

penguins [(penguins['island']== 'Torgersen') & (penguins['bill_length_mm']<35)]

```

```python
filter_penguin = penguins [(penguins['island']== 'Torgersen') | (penguins['bill_length_mm']<35)]
filter_penguin
```

```python
#quuery
#filter with .query ()
penguins.query ('island == "Torgersen" & bill_length_mm<35')     #"island = 'Togersan' "
```
